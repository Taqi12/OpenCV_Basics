
## 6. Rotation

### Simple Idea
Rotation means turning the image.

---

## 7. Why Rotate Images?

1. Fix alignment  
2. Increase dataset (Machine Learning)  
3. Improve training quality  

---

## 8. Rotation Syntax
cv2.rotate(src, rotateCode)

---

## 9. Rotation Types

### 90° Clockwise
cv2.ROTATE_90_CLOCKWISE  
→ Turn right  

### 180°
cv2.ROTATE_180  
→ Upside down  

### 90° Counterclockwise
cv2.ROTATE_90_COUNTERCLOCKWISE  
→ Turn left  

---

## 10. Rotation Examples

### 90° Right
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

### 180°
rotated = cv2.rotate(img, cv2.ROTATE_180)

### 90° Left
rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

---
## Final Summary


### Rotate
- Turn image (90°, 180°)
- Use cv2.rotate()

---

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("0_data/dog_img1.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# rotate clockwise by 90 degrees
rotate_90 = cv.rotate(img, cv.ROTATE_90_CLOCKWISE)

plt.imshow(rotate_90)
```

```python
# rotate counterclockwise by 90 degrees (-90 degrees)
rotate_neg_90 = cv.rotate(img, cv.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(rotate_neg_90)
```

```python
# rotate by 180 degrees
rotate_180 = cv.rotate(img, cv.ROTATE_180)

plt.imshow(rotate_180)
```
