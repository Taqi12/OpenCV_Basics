# Adding a border:
    - It means putting some extra pixels around the edges of your image. Think of it like framing a photo.

    - You can decide:

        - How wide the border is

        - What type of border you want

# How cv2.copyMakeBorder() Works

    - OpenCV takes your original image.

    - Makes a new bigger image.

    - Puts your original image in the center.

    - Fills the empty space around it with border pixels (depending on the type you choose).

# Border Types with Simple Examples
    - Constant Border (cv2.BORDER_CONSTANT)

        - Fills the border with a single color you choose.

        - Example: green frame around a picture

        - Layman Example: Like putting a colored frame around a painting.

    - Reflect Border (cv2.BORDER_REFLECT)

        - Mirrors the edge pixels.

        - Example: if the edge pixels are [10, 20, 30], the border becomes [30, 20, 10].

        - Layman Example: Like reflecting a photo in a mirror around its edges.

    - Replicate Border (cv2.BORDER_REPLICATE)

        - Repeats the edge pixels.

        - Example: if the edge pixel is 50, the border becomes [50, 50, 50, …].

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("0_data/dog_img1.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

border = cv.copyMakeBorder(img, 20,20,20,20, cv.BORDER_CONSTANT, value= (0,255,0))

plt.imshow(border)
```

```python
# Applying border around the image
border2 = cv.copyMakeBorder(img,100,100,200,200,cv.BORDER_REFLECT)
plt.imshow(border2)
```

```python
# Applying border around the image
border3 = cv.copyMakeBorder(img,100,100,200,200,cv.BORDER_REPLICATE)
plt.imshow(border3)
```
