# Advanced OpenCV Drawing Functions: 
    - Polylines, fillPoly, fillConvexPoly, ArrowedLine

#### cv2.polylines(image, [pts], isClosed, color, thickness)
    - Draws polylines/polygons from array of points `pts`. `isClosed=True` for closed shape.

#### cv2.fillPoly(img, [pts], color, lineType, shift, offset)
    - Fills one or more polygons defined by `pts` arrays with `color`.

#### cv2.fillConvexPoly(img, pts, color, lineType, shift)
    - Faster fill for convex polygons given `pts`; vertices point outwards, angles <180°.

#### cv2.arrowedLine(img, pt1, pt2, color, thickness, line_type, shift, tipLength)
    - Draws arrowed line from `pt1` to `pt2` with configurable tip length (default 0.3).

#### cv2.getTextSize(text, fontFace, fontScale, thickness)
    - Computes `(size, baseline)` for text bounding box, useful for drawing text backgrounds.

#### cv2.ellipse2Poly(center, axes, angle, arcStart, arcEnd, delta)
    - Approximates elliptic arc as polyline vertices (use with polylines for rendering).


```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((80, 128, 3), np.uint8)

plt.imshow(img)
```

```python
# Draw a circle in the image
cv.circle(img, center=(64, 40), radius=15, color=(255, 255, 255), thickness=-1)
plt.imshow(img)
```

```python
# Save the image
cv.imwrite('0_data/circle_image.png', img)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('0_data/circle_image.png')
plt.imshow(img)
```

```python
h, w = img.shape[:2]
#print(h, w)

# make points 
h1, w1 = int(h/2), int(w/2)

h2, w2 = int(h/4), int(w/4)

print("h1:", h1)
print("w1:", w1)
print("h2:", h2)
print("w2:", w2)
```

```python
# Draw something around the circle
points1 = np.array([[0, h1], [w1, h], [w, h1], [w1, 0]])

points2 = np.array([[0, h2], [w2, h], [w, h2], [w2, 0]])

cv.polylines(img, [points1], True, (0, 255, 0), 2)
cv.polylines(img, [points2], True, (0, 255, 0), 2)
plt.imshow(img)
```

```python
# Fill the poly
cv.fillPoly(img, [points1], (0, 255, 0))
cv.fillPoly(img, [points2], (0, 255, 0))
plt.imshow(img)
```

```python
# Draw arrowed line
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((80, 80, 3), dtype=np.uint8)

cv.arrowedLine(img, pt1=(10,50), pt2=(40,50), color=(255,0,255), thickness=1)
plt.imshow(img)
```

```python
# Text Size
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

font = cv.FONT_HERSHEY_SIMPLEX
font_scale = 1.0
font_thinkness = 1
bg_color = (255, 255, 255) # white
text_color = (0, 0, 0) # black

# text 
text = "Hello, World"

# Get box size
(text_width, text_height), baseline = cv.getTextSize(text, font, font_scale, font_thinkness)
text_patch = np.zeros((text_height + baseline, text_width, 3), np.uint8)
text_patch[:,:] = bg_color

cv.putText(text_patch, text, (0, text_height), font, font_scale, text_color, font_thinkness)
plt.imshow(text_patch)
```
