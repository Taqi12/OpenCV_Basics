# What is an Image Gradient?
    - Think of a gradient as the change in color or brightness in an image.

    - Example: If a pixel changes from dark to bright suddenly, that’s a high gradient.

    - Gradients tell us where the edges are in an image.

    - Layman Example:

        - Imagine a photo of a black cat on a white table.

            - The place where the cat’s body meets the table has a sudden change in brightness → that’s an edge → detected using gradients.

# Why Do We Need Gradients?

    - Gradients are very useful in computer vision:

    - Edge detection → Helps us see outlines of objects.

    - Feature extraction → Helps computers understand shapes.

    - Saliency maps → Highlight important parts of an image.

# How Edge Detection Works

    - Edges are where the pixel intensity changes sharply.

    - Gradients help measure how fast the intensity changes.

    - Big change = strong edge, small change = weak edge.

    - Layman Example:

        - If you’re coloring a cartoon and suddenly switch from black to white, the boundary is like an edge.

- Sobel Operator

    - What it does:

        - Calculates the gradient of the image in X (horizontal) and Y (vertical) directions.

        - Uses 3×3 kernels (small matrix) to approximate the derivative (change).

        - Horizontal kernel (Gx): detects vertical edges

            -1 0 1

            -2 0 2

            -1 0 1

        - Vertical kernel (Gy): detects horizontal edges

            -1 -2 -1

             0  0  0
             
             1  2  1

    - Combine Gx and Gy to get overall gradient:

        - Gradient = sqrt(Gx^2 + Gy^2)

        - Layman Example:

            - Imagine running a magnifying glass horizontally to see vertical lines → that’s horizontal gradient.

            - Running it vertically → see horizontal lines → vertical gradient.


# Scharr Operator

    - What it does:

        - Same as Sobel, but more accurate for detecting edges.

        - Works like Sobel but gives more weight to central pixels, so edges appear sharper.

        - Only works X or Y separately, not both.

        - Layman Example:

            - Imagine sharpening a pencil drawing to make lines darker and clearer → that’s like Scharr enhancing edges more than Sobel.

# Simple Analogy

    - Gradient :	How quickly brightness changes in a photo
    - Sobel :	A 3×3 magnifying glass looking for edges horizontally or vertically
    - Scharr :	A sharper magnifying glass for better edges
    
    - X-direction	Vertical lines (walls, fences)
    - Y-direction	Horizontal lines (tables, horizon)

- Sobel X → shows vertical edges

- Sobel Y → shows horizontal edges

- Scharr → same, but sharper

# Summary in Simple Words:

  - Gradient → measures how brightness changes in an image.

  - Sobel → basic edge detector, finds horizontal or vertical edges.

  - Scharr → improved Sobel, sharper edges, more accurate.

- Use case → detecting outlines of objects, contours, and shapes.

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("0_data/hor_ver_img.webp")
img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)

plt.imshow(img, cmap= 'gray')
```

```python
# X gradient Sobel
grad_x = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=5)
plt.imshow(grad_x, cmap='gray')
```

```python
# Y gradient Sobel
grad_y = cv.Sobel(img,cv.CV_64F,0,1,ksize=5)

plt.imshow(grad_y, cmap= 'gray')
```

```python
# X gradient Scharr
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("0_data/hor_ver_img.webp")
img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
 
scharr_x = cv.Scharr(img, cv.CV_32F, 1, 0) / 15.36

plt.imshow(scharr_x, cmap = "gray")

```

```python
# Y gradient Scharr
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("0_data/hor_ver_img.webp")
img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
 
scharr_y = cv.Scharr(img, cv.CV_32F, 0, 1) / 15.36

plt.imshow(scharr_y, cmap = "gray")

```
