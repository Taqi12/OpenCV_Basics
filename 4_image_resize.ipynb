# OpenCV Image Resizing and Rotation (Simple Notes)

## 1. Resize Function

### Syntax
cv2.resize(src, dsize, fx, fy, interpolation)

### Simple Idea
Resize means **changing the size of an image**.

---

## 2. Parameters Explanation

### src
- Input image (original image)

### dst
- Output image (result after resizing)
- Usually stored in a variable

### dsize
- Final size of image (width, height)
- Example: (300, 200)

### fx (Horizontal scaling)
- Controls width
- fx = 0.5 → half width  
- fx = 2 → double width  

### fy (Vertical scaling)
- Controls height
- fy = 0.5 → half height  
- fy = 2 → double height  

### Formula
new width  = original width × fx  
new height = original height × fy  

### interpolation
- Method to calculate new pixels
- Affects image quality

---

## 3. Interpolation Methods

### cv2.INTER_AREA
- Best for shrinking images
- Smooth result

### cv2.INTER_LINEAR
- Default method
- Good for normal resizing

### cv2.INTER_CUBIC
- Best for enlarging images
- High quality but slower

---

## 4. Using (0, 0) in Resize

### Example
resized = cv2.resize(img, (0,0), fx=0.5, fy=0.5)

### Meaning
- (0,0) means: do not use fixed size
- OpenCV will calculate size using fx and fy

### Important Formula
new width  = original width × fx  
new height = original height × fy  

### Why use (0,0)?
- Easier when scaling
- No need to calculate size manually

---

## 5. Two Ways to Resize

### Method 1: Using fixed size
cv2.resize(img, (300, 200))

### Method 2: Using scaling
cv2.resize(img, (0,0), fx=0.5, fy=0.5)

---

## Final Summary

### Resize
- Change image size
- Use dsize OR fx, fy
- Choose interpolation for quality


## One-Line Key Concept

(0,0) in resize = Ignore size, use scaling factors (fx, fy)

```python
import cv2 as cv
import matplotlib.pyplot as plt

# original image
img = cv.imread('0_data/dog_img1.jpg')
# convret bgr to rgb
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# Half the image
half_image = cv.resize(img, (0, 0), fx=0.5, fy=0.5)

plt.imshow(half_image)
```

```python
# Double the image

double_image = cv.resize(img, (0, 0), fx=2.0, fy=2.0)
plt.imshow(double_image)
```

```python
# Resize the image to a specific width and height
resized_image = cv.resize(img, (100, 100))
plt.imshow(resized_image)
```

```python
# Resizing with interpolation
area_resize = cv.resize(img, (0, 0), fx=0.5, fy=0.5, interpolation=cv.INTER_LINEAR)
cubic_resize = cv.resize(img, (0, 0), fx=0.5, fy=0.5, interpolation=cv.INTER_CUBIC)
linear_resize = cv.resize(img, (0, 0), fx=0.5, fy=0.5, interpolation=cv.INTER_LINEAR)

name = ["Original", "Area", "Cubic", "Linear"]

images = [img, area_resize, cubic_resize, linear_resize]

plt.figure(figsize=(10, 10))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i])
    plt.title(name[i])
    #plt.axis("off")
plt.show()
```
