# What is Blurring?

Blurring means smoothing an image by reducing sharp changes in pixel intensity.

# Why Blurring is Important

    - Remove noise:
      - Tiny specks of wrong colors in the image get averaged out.

    - Remove high-frequency content:
      - Sharp edges or tiny details that aren’t important can be softened.

    - Better feature recognition:
      - Reducing noise helps algorithms like edge detection or face detection work better.

## Average Blur
    - How: Takes the average of all neighbor pixels and replaces the center pixel.
  
    - Think: “Mix all colors in the neighborhood evenly.”

    - Effect: Smooths the image, edges get blurry.

    - Use: To remove small, random noise.

## Gaussian Blur

    - How: Takes a weighted average, giving more importance to center pixels.

    - Think: “Mix colors, but focus more on the center.”

    - Effect: Smooths image, edges stay sharper than average blur.

    - Use: To remove normal noise, gentle smoothing.

## Median Blur

    - How: Replaces the center pixel with the median (middle value) of neighbors.

    - Think: “Ignore extreme spots, pick the middle color.”

    - Effect: Removes salt-and-pepper noise (random black/white dots) without blurring edges.

    - Use: When the image has random black/white dots.


```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread("0_data/dog_img1.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
img.shape
```

```python
# Averaging the image
blur_img = cv.blur(img, (5, 5))
plt.imshow(blur_img)
```

- Blur only change the pixels values not the shape of the image

```python
# pixels of original image at point row 100, column 150
org_pixels = img[100, 150]
print("Original pixels:", org_pixels)
# pixels of blur image
blur_pixels = blur_img[100, 150]
print("Blur pixels:", blur_pixels)
```

```python
# Gaussian blur
gauss_img = cv.GaussianBlur(img, (5, 5), 0)
plt.imshow(gauss_img)
```

```python
# pixels of original image at point row 100, column 150
org_pixels = img[100, 150]
print("Original pixels:", org_pixels)
# pixels of Gaussian blur image
gauss_pixels = gauss_img[100, 150]
print("Gaussian blur pixels:", gauss_pixels)
```

```python
# Median Blur
median_img = cv.medianBlur(img, 5)

plt.imshow(median_img)
```

```python
# pixels of original image at point row 100, column 150
org_pixels = img[100, 150]
print("Original pixels:", org_pixels)
# pixels of Median blur image
median_pixels = median_img[100, 150]
print("Median blur pixels:", median_pixels)
```
