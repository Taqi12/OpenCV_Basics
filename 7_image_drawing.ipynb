## Drawing Functions in OpenCV

### Create Line in OpenCV using `cv2.line()`

In geometry, a part of a line that is bounded by two distinct endpoints, and contains every point between its endpoints on the line is called a line segment.

This function is used to draw a line segment that connects any two points. It draws a line between points `p1` and points `p2`. The line gets clipped by the boundaries of the image.

**Syntax:**

```python
cv2.line(img, pt1, pt2, color, thickness, lineType, shift)
```

**Parameters:**

- `img`: Source image or image over which we have to draw the line
- `pt1`: Starting coordinates of the line segment.
- `pt2`: Ending coordinates of the line segment
- `color`: The color of the line
- `thickness`: Thickness of the line
- `lineType`: Type of the line offered by OpenCV
- `shift`: Fractional bits in the point coordinates

### Create Rectangle in OpenCV using `cv2.rectangle()`

This function is used to create a rectangle in OpenCV. A rectangle can be defined as a plane figure with four straight sides and right angles, and unequal adjacent sides. The opposite corners of the rectangle represent the two points `pt1` and `pt2`.

**Syntax:**
```python
cv2.rectangle(img, pt1, pt2, color, thickness, lineType, shift)
```

**Parameters:**
- `img`: Source image or image over which we have to draw the line
- `pt1`: Starting coordinate of the rectangle
- `pt2`: Ending coordinate of the triangle on the opposite
- `color`: Color of the rectangle
- `thickness`: Thickness of the lines of the rectangle. Negative thickness refers to a filled rectangle
- `lineType`: Type of the line offered by OpenCV
- `shift`: Fractional bits in the point coordinates

### Create clipped segment in OpenCV using `cv2.clipLine()`

The `cv2.clipLine()` function returns a clipped segment defined inside the rectangle. If both the points of the segment, `pt1` and `pt2` are outside the rectangle, then the return value is `False`, otherwise, if some of the points of the clip segment are inside the rectangle, then the return value is `true`.

**Syntax:**
```python
cv2.clipLine(imgSize, pt1, pt2)
```

**Parameters:**
- `imgSize`: Size of the image. The rectangle is defined as `rect( 0, 0, imgSize.width, imgSize.height)`
- `pt1`: Starting coordinate for the clip line
- `pt2`: Ending coordinate for the clip line


### Create circle in OpenCV using `cv2.circle()`

A circle consists of all points in a plane that are at a given distance from the center and equivalently it is the curve traced out by a point that moves in a plane with the distance from the center remaining constant. The distance from the center to the traced curve is termed as the radius of the circle. The `cv2.circle` function is used to draw a circle of the given radius. The center of the circle refers to the coordinate of the window where the center of the circle will be drawn.

**Syntax:**
```python
cv2.circle(img, center, radius, color, thickness, lineType, shift)
```

**Parameters:**
- `img`: Source image or image over which we have to draw the line
- `center`: Coordinate where the center of the circle will be drawn
- `radius`: Radius of the circle
- `color`: Color of the circle
- `thickness`: Thickness of the lines of the circle. Negative thickness refers to a filled circle.
- `lineType`: Type of the line offered by OpenCV
- `shift`: Fractional bits in the point coordinates

### Create ellipse in OpenCV using `cv2.ellipse()` function

This function is used to draw an ellipse outline, a filled ellipse, or an arc over any image. For this function, we need to specify the major and minor axis of the ellipse, the angle of rotation, and the starting and ending angles. These parameters are specified to draw an arc.

We can create an ellipse with horizontal and vertical orientation.

**Syntax:**
```python
cv2.ellipse(src, center, axes, angle, startAngle, endAngle, color, thickness, lineType, shift)
```

**Parameters:**
- `src`: Source image or image over which we have to draw the ellipse
- `center`: Coordinate where the center of the ellipse will be drawn. It is represented by a tuple.
- `axes`: Tuple containing the length of the major and minor axes.
- `angle`: Rotation angle
- `startAngle`: Starting angle
- `endAngle`: Ending angle
- `color`: Color of the border of the ellipse
- `thickness`: Thickness of the border
- `lineType`: Type of line. It is an optional parameter.
- `shift`: Fractional bits in the point coordinates

### Display Text data in OpenCV using `cv2.putText()`

The function `cv2.putText` displays the specified text data over an image in OpenCV. To put text over images, we need to specify the text data, the position of the coordinates of the image where the text will be displayed, font type and scale of the text data, and other parameters such as color, thickness, and lineType.

**Syntax:**
```python
cv2.putText(src, text, org, font, scale, color, thickness, lineType, bottomLeftOrigin)
```

**Parameters:**
- `src`: Source image or input image over which we'll write
- `text`: Text to be written
- `org`: Specified as the bottom-left corner of the text
- `font`: Font of the text
- `scale`: Scale factor to be multiplied with the base size of a font
- `color`: Color of the text
- `thickness`: The thickness of the text
- `lineType`: Type of line
- `bottomLeftOrigin`: It specifies the org to be bottom-left corner if true, otherwise, top-left corner


````python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((128, 128, 3), np.uint8)
plt.imshow(img)
````

````python
# Draw a line
# color = (B, G, R)
cv.line(img, pt1 = (10, 10), pt2 = (50, 50), color = (0, 255, 0), thickness=2)
plt.imshow(img)
````

````python
# Draw a rectangle

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((128, 128, 3), np.uint8)

# rectangle
cv.rectangle(img, pt1=(10, 10), pt2=(50, 50), color=(255, 0, 0), thickness=2)

plt.imshow(img)

````

````python
# Clipline : Return the part of a segment
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((128, 128, 3), np.uint8)

# Creating a rectangle using the cv2.rectangle function
cv.rectangle(img, (5, 5), (50, 50), (0,255,0), 2)

# Clipline
rect, pt1, pt2 = cv.clipLine((0, 0, 128, 128), (10, 10), (50, 50))
if rect:
    cv.line(img, pt1, pt2, (255, 0, 0), 2)

plt.imshow(img)
````

````python
# Circle
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((128, 128, 3), np.uint8)

cv.circle(img, center= (50, 50), radius= 20, color= (0, 0, 255), thickness= 1)

plt.imshow(img)
````

````python
# Creating a filled circle
cv.circle(img, center= (50, 50), radius= 20, color= (0, 0, 255), thickness= -1)
plt.imshow(img)
````

````python
# Eclipse
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((128, 128, 3), np.uint8)

cv.ellipse(img, center=(50, 50), axes=(30, 20), angle=0, startAngle=0, endAngle=360, color=(255, 0, 0), thickness=2)
plt.imshow(img)
````

````python
# Write text in image
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((128, 128, 3), np.uint8)

font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img, 'Hello, World!', org=(10, 50), fontFace=font, fontScale=0.5, color=(255, 255, 255), thickness=1)
plt.imshow(img)
````
