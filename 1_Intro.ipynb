```python
import cv2
import matplotlib.pyplot as plt

# it will show the bgr = blue green red image
img = cv2.imread('0_data/dog_img1.jpg')

plt.imshow(img)
```

```python
# converting bgr to rgb
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
plt.imshow(new_img)
```

```python
# Flip the image
flipped_img_horizontally  = cv2.flip(new_img, 1)  # 1 for horizontal flip, 0 for vertical flip
plt.imshow(flipped_img_horizontally)
```

```python
# Flip image vertically
flipped_img_vertically = cv2.flip(new_img, 0)  

plt.imshow(flipped_img_vertically)
```

```python
# Flip image 
flipped_img_vertically = cv2.flip(new_img, -1)  

plt.imshow(flipped_img_vertically)
```

```python
# Flip image 
flipped_img_vertically = cv2.flip(new_img, -0)  

plt.imshow(flipped_img_vertically)
```
